# Final Organized Dataset Analysis

### Import Necessary Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
# Define the path to the FinalDataset
final_data_dir = "./FinalDataset"

# Initialize lists to store data
data = []
labels = []

# Loop through the subdirectories (labels) in the FinalDataset
for label in os.listdir(final_data_dir):
    label_dir = os.path.join(final_data_dir, label)
    if os.path.isdir(label_dir):
        for filename in os.listdir(label_dir):
            if filename.endswith(".jpg"):
                img_path = os.path.join(label_dir, filename)
                labels.append(label)
                data.append(img_path)

# Create a DataFrame for analysis
df = pd.DataFrame({"Image_Path": data, "Label": labels})
```

### Basic Dataset analysis

```python
# Basic data analysis
label_counts = df["Label"].value_counts()
total_images = len(df)

print("Number of Images:", total_images)
print("Number of Labels:", len(label_counts))
print("\nLabel Distribution:")
print(label_counts)
```

### Plot label distribution

```python
# Plot label distribution
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.countplot(data=df, y="Label", order=label_counts.index)
plt.title("Label Distribution")
plt.xlabel("Count")
plt.ylabel("Label")
plt.show()
```

### Display some sample images for each label

```python
# Display some sample images for each label
plt.figure(figsize=(15, 5))
i = 1
sample_images = df.groupby("Label").first()
for index, row in sample_images.iterrows():
    img = plt.imread(row["Image_Path"])
    plt.subplot(1, 3, i)
    plt.imshow(img)
    plt.title(index)
    plt.axis("off")
    i += 1

plt.show()
```
